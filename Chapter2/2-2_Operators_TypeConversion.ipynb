### Basic Operators in Python

In Python, operators are used to perform operations on variables and values. These operators are divided into several categories:<br>
- Artihmatic Operators:
    - `+` Addition
    - `-` Subtraction
    - `*` Multiplication
    - `/` Division
    - `%` Modulus (remainder)
    - `**` Exponentiation
    - `//` Floor Division<br><br>
    


```python
# Addition
a = 10
b = 5
result = a + b
result
```

```python
# Subtraction
result = a - b 
result
```

```python
# Multiplication
result = a * b
result
```

```python
# Division
result = a / b
result
```

**Note**: Divsion will always result in a `float`.

```python
# Modulus
result = a % b
result
```

A modulus operator returns the remaining portion of a division problem.  Since 5 goes into 10 evenly (2 times), the result will be 0.

```python
# Exponentiation
result = a ** 2 
result
```

```python
# Floor Division
result = a // b
result
```

Floor division in Python will only return the number of times the right operand goes into the left operand evenly, leaving out any remainder.

- Comparison Operators:
    - `==` Equal to
    - `!=` Not Equal to
    - `>` Greater than
    - `<` Less than
    - `>=` Greater than or equal to
    - `<=` Less than or equal to <br><br>




```python
# Equals to
result = (a == b)
result
```

```python
# Not equal to
result = (a != b)
result
```

```python
# Greater than
result = (a > b)
result
```

```python
# Less than
result = (a < b)
result
```

```python
# Greater than or equal to
result = (a >= b)
result
```

```python
# Less than or equal to
result = (a <= b)
result
```

- Logical Operators:
    - `and`/`&`
    - `or`/`|`
    - `not` <br><br>

```python
# and/&
result_and1 = (a > 0) and (b > 0)
result_and2 = (a > 0) & (b > 0)

print(result_and1)
print(result_and2)
```

Here, `and` and `&` can be used interchangably.

```python
# or/|
result_or1 = (a > 11) or (b > 0)
result_or2 = (a > 12) | (b < -1)

print(result_or1)
print(result_or2)
```

Here, `or` and `|` can be used interchangably.

```python
result1 = not(a > 0)
result2 = not(a < 0)

print(result1)
print(result2)
```

- Assignment Operators:
    - `=` Assigns the value of the right operand to the left operand
    - `+=` Adds the right operand to the left operand
    - `-=` Subtracts the right operand from the left operand
    - `*=` Multiplies the right operand with the left operand
    - `/=` Divides the left operand by the right operand
    - etc. for all other arithmatic operators <br><br>



```python
# =
c = a + b 
c
```

```python
# +=
a = 10
b = 5

a += b
a
```

```python
# -=
a = 10
b = 5

a -= b
a
```

```python
# *=
a = 10
b = 5

a *= b
a
```

```python
# /=
a = 10
b = 5

a /= b
a
```

```python
# %=
a = 10
b = 5

a %= b
a
```

```python
# **=
a = 10
b = 5

a **= b
a
```

```python
# //=
a = 10
b = 5

a //= b
a
```

- Identity Operators:
    - `is` True if both operands refer to the same object
    - `is not` True is operands refer to different objects<br><br>




```python
x = [1, 2, 3]
y = x
result = (x is y)
result
```

```python
z = [1, 2, 3]
result = (x is not z)
result
```

- Membership Operators:
    - `in` True if the value is found in the sequence of values
    - `not in` True is the value is not found in th esequence of values<br><br>

```python
result = (5 in [1, 2, 3, 4, 5])
result
```

```python
result = (6 not in [1, 2, 3, 4, 5])
result
```

These operators form the building blocks of Python programming, allowing for the manipulation and comparison of data in powerful and flexible ways.

### Type Conversion

Type conversion in Python is about converting one data type into another. 

This is often necessary in Python programming, as you may need to manipulate values in a different format than they were originally in. 

Python provides several built-in functions for explicit type conversion:<br>


- `int()` Converts a value to an integer. Useful for converting floats or strings representing integers into integer values.





```python
num_int = int(3.14) # converts `3.14` to `3`
num_int
```

Here, in converting 3.14 to an integer, it loses its decimal part, returning the whole number remaining.


- `float()` Converts a value to a floating-point number. Used for converting integers or strings representing floating-point values into float.




```python
num_float = float("3.14") # converts the string `"3.14"` to `3.14`
num_float
```

Here, the result <i>looks</i> like it could be a string, but when you use the `type()` function, it shows that it is, indeed, a float:

```python
type(num_float)
```

- `str()` Converts a value to a string. It's often used for concatenating numbers with strings or formatting.

```python
str_val = str(10) # converts the integer `10` to the string `"10"
print(str_val, type(str_val))
```

- `list()` Converts a value to a list. Useful for turning tuples, strings, or sets into lists.

```python
list_val = list((1, 2, 3)) # converts the tuple `(1, 2, 3)` to a list `[1, 2, 3]`
print(list_val, type(list_val))
```

- `tuple()` Converts a value to a tuple. Useful for making lists or other iterables immutable.

```python
tuple_val = tuple([1, 2, 3]) # converts the list `[1, 2, 3]` to a tuple `(1, 2, 3)`
print(tuple_val, type(tuple_val))
```

- `set()` Converts a value to a set. Often used to remove duplicates from a list or tuple.
   - `set_val = set([1, 1, 2, 3])` converts the list `[1, 1, 2, 3]` to a set `{1, 2, 3}`<br><br>



- `dict()` Converts a list of pairs (or other iterable of pairs) to a dictionary.

```python
dict_val = dict([('key1', 'value1'), ('key2', 'value2')]) # converts a list of pairs to a dictionary `{'key1': 'value1', 'key2': 'value2'}`.
print(dict_val, type(dict_val))
```
